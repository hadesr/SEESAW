```python
import numpy as np
import matplotlib.pyplot as plt
# PARAMS TO COMPARE
# ipc, dtlb hit rate, stlb hit rate, l1d hit rate, 
```

```python
############################################# ipc #############################################
X = ['bellmann','bfs','components','mis','sssp']
baseline = np.array([0.119322,0.151018,0.104598,0.072088,0.205711])
all_sp_1 = np.array([0.129173,0.145315,0.108781,0.071832,0.208636])
all_sp_2 = np.array([0.124256,0.143431,0.105588,0.0689921,0.189981])

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.25, baseline, 0.25, label = 'baseline')
plt.bar(X_axis - 0.0, all_sp_1, 0.25, label = 'superpages')
plt.bar(X_axis + 0.25, all_sp_2, 0.25, label = 'partition')

plt.xticks(X_axis, X)
plt.xlabel("Traces")
plt.ylabel("IPC")
plt.title("IPC for each trace and settings")
plt.legend(loc=9, ncol=3)
ax = plt.gca()
ax.set_ylim([0.0, 0.25])
plt.savefig('ipc.jpg')
plt.show()
```
